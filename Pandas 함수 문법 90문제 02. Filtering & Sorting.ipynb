### Question 33

df의 item_name 컬럼 값이 Steak Salad 또는 Bowl 인 데이터를 데이터 프레임화 한 후,  
item_name를 기준으로 중복행이 있으면 제거하되 첫번째 케이스만 남겨라

```python
q33 = df.loc[(df['item_name'] == 'Steak Salad') | (df['item_name'] == 'Bowl')]
q33 = q33.drop_duplicates(['item_name'], keep='first')
q33.head()
```

### Question 34
