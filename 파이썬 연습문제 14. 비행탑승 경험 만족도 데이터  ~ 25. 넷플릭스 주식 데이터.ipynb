# 14. 비행탑승 경험 만족도 데이터

> Attention
- 데이터 출처 :https://www.kaggle.com/teejmahal20/airline-passenger-satisfaction?select=train.csv (참고, 데이터 수정)
- 데이터 설명 : 비행탑승 경험 만족도 (satisfaction 컬럼 : ‘neutral or dissatisfied’ or satisfied ) (83123, 24) shape
- Data url :https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/train.csv

```python
import pandas as pd

df = pd.read_csv("https://raw.githubusercontent.com/Datamanim/datarepo/main/airline/train.csv")
df.head()
```

### Question 1

Arrival Delay in Minutes 컬럼이 결측치인 데이터들 중 ‘neutral or dissatisfied’ 보다 ‘satisfied’의 수가 더 높은 Class는 어디 인가?

```python
q1 = df[df['Arrival Delay in Minutes'].isnull()].groupby(['Class', 'satisfaction']).size().reset_index()
q1 = q1.pivot_table(index='Class', columns='satisfaction')

q1[q1[0]['satisfied'] > q1[0]['neutral or dissatisfied']].index
```

> **Answer**

```python
answer =df.loc[df['Arrival Delay in Minutes'].isnull()].groupby(['Class','satisfaction'],as_index=False).size().pivot(index='Class',columns='satisfaction')
result =answer[answer['size']['neutral or dissatisfied'] < answer['size']['satisfied']]
result
```

### Question 3

흡연자와 비흡연자 각각 charges의 상위 10% 그룹의 평균의 차이는?

```python
smoke_high = df[df['smoker'] == 'yes']['charges'].quantile(0.9)
nosmoke_high = df[df['smoker'] == 'no']['charges'].quantile(0.9)

smoke_high_mean = df[(df['smoker'] == 'yes') & (df['charges'] >= smoke_high)]['charges'].mean()
nosmoke_high_mean = df[(df['smoker'] == 'no') & (df['charges'] >= nosmoke_high)]['charges'].mean()

smoke_high_mean - nosmoke_high_mean
```

> **Answer**

```python
high = df.loc[df.smoker =='yes'].charges.quantile(0.9)
high2 = df.loc[df.smoker =='no'].charges.quantile(0.9)

mean_yes = df.loc[(df.smoker =='yes') &(df.charges >=high)].charges.mean()
mean_no = df.loc[(df.smoker =='no') &(df.charges >=high2)].charges.mean()

answer = mean_yes - mean_no
print(answer)
```
