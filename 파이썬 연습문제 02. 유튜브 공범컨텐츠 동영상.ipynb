### Question 3

### Question 4

각채널의 2021-10-03 03:00:00 ~ 2021-11-01 15:00:00 까지 구독자수 (subcnt) 의 증가량을 구하여라

```python
q4 = channel.loc[(channel['ct'] > pd.to_datetime('2021-10-03 03:00:00')) & (channel['ct'] < pd.to_datetime('2021-11-01 15:00:00'))]
q4
```

```python
print(q4['channelname'].nunique())
print(q4['channelname'].unique())
```

```python
q4_name = list(q4['channelname'].unique())
q4_name
```

```python
q4_list = []

for c in q4_name:
    q4_1 = q4.loc[q4['channelname'] == c].sort_values('ct').reset_index(drop=True)
    increase = int(q4_1.subcnt.tail(n=1) - q4_1.subcnt[0])
    q4_list.append([c, increase])
```

### Question 7

“공범” 컨텐츠의 경우 19:00시에 공개 되는것으로 알려져있다.  
공개된 날의 21시의 viewcnt, ct, videoname 으로 구성된 데이터 프레임을 viewcnt를 내림차순으로 정렬하여 출력하라

```python
q7 = video[['videoname','viewcnt','ct']].loc[video['ct'].dt.hour == 21].drop_duplicates('videoname').sort_values('ct', ascending=False).reset_index(drop=True)
q7
```

> **Answer**

```python
video['time']= video.ct.dt.hour

answer = video.loc[video['time'] ==21] \
            .sort_values(['videoname','ct'])\
            .drop_duplicates('videoname') \
            .sort_values('viewcnt',ascending=False)[['videoname','viewcnt','ct']]\
            .reset_index(drop=True)

display(answer)
```
