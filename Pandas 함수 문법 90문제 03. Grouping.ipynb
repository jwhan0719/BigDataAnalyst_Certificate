### Question 44

데이터를 로드하고 상위 5개 컬럼을 출력하라

```python
DataUrl = 'https://raw.githubusercontent.com/Datamanim/pandas/main/AB_NYC_2019.csv'
df = pd.read_csv(DataUrl)

df.head()
```

### Question 45

데이터의 각 host_name의 빈도수를 구하고 host_name으로 정렬하여 상위 5개를 출력하라

```python
q45 = df['host_name'].value_counts()
q45.head(5)
```

### Question 46

데이터의 각 host_name의 빈도수를 구하고 빈도수 기준 내림차순 정렬한 데이터 프레임을 만들어라. 빈도수 컬럼은 counts로 명명하라

```python
q46 = pd.DataFrame(df.groupby('host_name').size().sort_values(ascending=False)).rename(columns={0:'counts'})
q46.head(5)
```

### Question 47

neighbourhood_group의 값에 따른 neighbourhood컬럼 값의 갯수를 구하여라

```python
q47 = df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size()
q47.head()
```

### Question 48

neighbourhood_group의 값에 따른 neighbourhood컬럼 값 중 neighbourhood_group그룹의 최댓값들을 출력하라

```python
q48 = df.groupby(['neighbourhood_group', 'neighbourhood'], as_index=False).size().groupby(['neighbourhood_group'], as_index=False).max()
q48.head()
```

### Question 49

neighbourhood_group 값에 따른 price값의 평균, 분산, 최대, 최소 값을 구하여라

```python
q49 = df[['neighbourhood_group', 'price']].groupby('neighbourhood_group').agg(['mean', 'var', 'max', 'min'])
q49
```

### Question 50

neighbourhood_group 값에 따른 reviews_per_month 평균, 분산, 최대, 최소 값을 구하여라

```python
q50 = df[['neighbourhood_group', 'reviews_per_month']].groupby('neighbourhood_group').agg(['mean', 'var', 'min', 'max'])
q50.head()
```

### Question 51

neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 구하라

```python
q51 = pd.DataFrame(df.groupby(['neighbourhood', 'neighbourhood_group'], as_index=False).price.mean())
q51.head()
```

### Question 52

neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하라

```python
q52 = df.groupby(['neighbourhood','neighbourhood_group']).price.mean().unstack()
q52.head()
```

### Question 53

neighbourhood 값과 neighbourhood_group 값에 따른 price 의 평균을 계층적 indexing 없이 구하고 nan 값은 -999값으로 채워라

```python
q53 = df.groupby(['neighbourhood','neighbourhood_group']).price.mean().unstack().fillna(-999)
q53.head()
```

### Question 54

데이터중 neighbourhood_group 값이 Queens값을 가지는 데이터들 중 neighbourhood 그룹별로 price값의 평균, 분산, 최대, 최소값을 구하라

```python
q54 = df.loc[df['neighbourhood_group'] == 'Queens'].groupby('neighbourhood').price.agg(['mean', 'var', 'max', 'min'])
q54.head()
```

### Question 55

데이터중 neighbourhood_group 값에 따른 room_type 컬럼의 숫자를 구하고 neighbourhood_group 값을 기준으로 각 값의 비율을 구하여라

```python
q55 = df.groupby(['neighbourhood_group', 'room_type']).size().unstack()
q55.loc[:,:] = (q55.values /q55.sum(axis=1).values.reshape(-1,1))
q55
```
