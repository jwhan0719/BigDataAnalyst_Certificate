## T1-21. Binning Data / #비닝 #구간나누기

### 21.  구간 분리
- 데이터셋 : basic1.csv
- 'age'컬럼 이상치를 제거하고, 동일한 개수로 나이 순으로 3그룹으로 나눈 뒤 각 그룹의 age 컬럼의 중앙값을 더하시오
- (이상치는 음수(0포함), 소수점 값)
- https://www.kaggle.com/code/agileteam/py-t1-21-expected-question/notebook

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/basic1.csv')

display(df.head(), df.shape)
```

```python
df['group'] = pd.qcut(df['age'], q=3, labels=['grp1', 'grp2', 'grp3'])
```

```python
np.sort(df['age'].unique())
```

```python
df = df[(df['age'] > 0) & (df['age'] == round(df['age'], 0))]
df
```

```python
np.sort(df['age'].unique())
```

```python
df['group'] = pd.qcut(df['age'], q=3, labels=['grp1', 'grp2', 'grp3'])

df['group'].value_counts()
```

```python
grp1 = df[df['group'] == 'grp1']['age'].median()
grp2 = df[df['group'] == 'grp2']['age'].median()
grp3 = df[df['group'] == 'grp3']['age'].median()
total_grp = grp1+grp2+grp3

print('그룹1 중앙값 :', grp1, '\n그룹2 중앙값 :', grp2, '\n그룹3 중앙값 :', grp3, '\n합계 :', total_grp)
```

## T1-22. Time-Series4 (Weekly data) / #주간 #합계

- 주 단위 Sales의 합계를 구하고, 가장 큰 값을 가진 주와 작은 값을 가진 주의 차이를 구하시오(절대값)

```python
low = df['Sales'].min()
high = df['Sales'].max()

abs(high-low)
```

## T1-23. Drop Duplicates / #중복제거 #결측치 #10번째값으로채움

### 23.  중복 제거
- 데이터셋 : basic1.csv
- f1의 결측치를 채운 후 age 컬럼의 중복 제거 전과 후의 'f1' 중앙값 차이를 구하시오
- 결측치는 f1의 데이터 중 내림차순 정렬 후 10번째 값으로 채움
- 중복 데이터 발생시 뒤에 나오는 데이터를 삭제함
- 최종 결과값은 절대값으로 출력
- https://www.kaggle.com/code/agileteam/t1-23-drop-duplicates/notebook

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/basic1.csv')

display(df.head(), df.shape)
```

```python
f1_null = df[df['f1'].isnull()]['f1'].index
df.loc[f1_null]['f1'].values
```

```python
top10_value = df['f1'].sort_values(ascending=False).values[9]
df['f1'] = df['f1'].fillna(top10_value)

top10_value, df.loc[f1_null]['f1'].values
```

```python
bf_age = df['f1'].median()

drop_df = df.drop_duplicates('age')
af_age = drop_df['f1'].median()

total_age = abs(bf_age - af_age)

print('중복 제거전 :', df.shape, '\n중복 제거후 :', drop_df.shape)
print('\n중복 제거전 값 :', bf_age, '\n중복 제거후 값 :', af_age, '\n값 차이 :', total_age)
```

## T1-24. Time-Series5 (Lagged Feature) / #시차데이터 #조건

```python
df[(df['Events'] == 1) & (df['Sales'] <= 1000000)]['PV2'].sum()
```
