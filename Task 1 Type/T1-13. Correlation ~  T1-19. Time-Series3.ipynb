## T1-13. Correlation / #상관관계

### 13. 상관관계
- 데이터셋 : winequality-red.csv
- 주어진 데이터에서 상관관계를 구하고, quality와의 상관관계가 가장 큰 값과, 가장 작은 값을 구한 다음 더하시오
- 단, quality와 quality 상관관계 제외, 소수점 둘째 자리까지 출력
- https://www.kaggle.com/code/agileteam/py-t1-13-expected-questions/notebook

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/winequality-red.csv')

display(df.head())
df.shape
```

```python
corr = df.corr()
corr = corr[corr['quality'] < 1]

high = corr['quality'].max()
low = corr['quality'].min()
total = round(high+low, 2)

print('상관관계 가장 큰 값 :', high, '\n상관관계 가장 작은 값 :', low, '\n합계 :', total)
```

## T1-16. Variance / #분산 #결측치전후값차이

### 16. 분산
- 데이터셋 : basic1.csv
- f2가 0값인 데이터를 age를 기준으로 오름차순 정렬하고 앞에서 부터 20개의 데이터를 추출한 후 
- f1 결측치(최소값)를 채우기 전과 후의 분산 차이를 계산하시오 (소수점 둘째 자리까지)
- https://www.kaggle.com/code/agileteam/py-t1-16-expected-question/notebook

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/basic1.csv')

display(df.head())
df.shape
```

```python
top20 = df[df['f2'] == 0].sort_values(by='age').head(20)
top20
```

```python
bf_var = top20['f1'].var()

top20['f1'] = top20['f1'].fillna(top20['f1'].min())
af_var = top20['f1'].var()

total_var = round(bf_var-af_var, 2)

print('결측치 대체전 분산 값 :', bf_var, '\n결측치 대체후 분산 값 :', af_var, '\n분산 차이 :', total_var)
```

## T1-18. Time-Series2 / #주말 #평일 #비교 #시계열데이터

### 18. 시계열2
- 데이터셋 : basic2.csv
- 주어진 데이터에서 2022년 5월 주말과 평일의 sales컬럼 평균값 차이를 구하시오 (소수점 둘째자리까지 출력, 반올림)
- https://www.kaggle.com/code/agileteam/py-t1-18-2-expected-question/notebook

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/basic2.csv')

display(df.head())
df.shape
```

```python
df['Date'] = pd.to_datetime(df['Date'])
df['Week'] = df['Date'].dt.dayofweek # 0-월요일 ~ 6-일요일

q18 = df[(df['Date'].dt.year == 2022) & (df['Date'].dt.month == 5)]

weekend = q18[q18['Week'].isin([5,6])]['Sales'].mean()
weekday = q18[~q18['Week'].isin([5,6])]['Sales'].mean()
week_total = round(weekend-weekday, 2)

print('주말 평균값 :', weekend, '\n평일 평균값 :', weekday, '\n평균값 차이 :', week_total)
```

## T1-19. Time-Series3 (monthly total) / #월별 #총계 #비교 #데이터값변경

### 19. 시계열3
- 데이터셋 : basic2.csv
- 2022년 월별 Sales 합계 중 가장 큰 금액과 2023년 월별 Sales 합계 중 가장 큰 금액의 차이를 절대값으로 구하시오.
- 단 Events컬럼이 '1'인경우 80%의 Salse값만 반영(최종값은 소수점 반올림 후 정수 출력)
- https://www.kaggle.com/code/agileteam/py-t1-19-3-expected-question/notebook

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/basic2.csv')

display(df.head())
df.shape
```

```python
def q19(row):
    if row['Events'] == 1:
        return row['Sales']*0.8
    else:
        return row['Sales']
```

```python
df['Sales'] = df.apply(q19, axis=1)

df['Date'] = pd.to_datetime(df['Date'])
df['year'] = df['Date'].dt.year
df['month'] = df['Date'].dt.month

df
```

```python
df_2022 = df[(df['year'] == 2022)].groupby('month')['Sales'].sum().max()
df_2023 = df[(df['year'] == 2023)].groupby('month')['Sales'].sum().max()
df_total = round(abs(df_2022 - df_2023), 0)

print('2022년 최고금액 :', df_2022, '\n2023년 최고금액 :', df_2023, '\n금액 차이 :',df_total)
```
