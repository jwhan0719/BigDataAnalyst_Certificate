## T1-1.Outlier(IQR) / #이상치 #IQR

### 1. 이상치를 찾아라
- 데이터셋 : titanic_train.csv
- IQR을 활용해 Fare 컬럼 이상치 찾고, 이상치 데이터의 여성의 수 구하기
- https://www.kaggle.com/code/agileteam/py-t1-1-iqr-expected-questions/notebook

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/titanic_train.csv')

df.head(3)
```

```python
df.isnull().sum()
```

```python
q1 = df['Fare'].quantile(0.25)
q3 = df['Fare'].quantile(0.75)

iqr = q3-q1
cut_off = 1.5*iqr

lower = q1-cut_off
upper = q3+cut_off

lower, upper
```

```python
print('Fare 컬럼 이상치 개수 :', len(df[(df['Fare'] < lower) | (df['Fare'] > upper)]))

out_df = df[(df['Fare'] < lower) | (df['Fare'] > upper)]
print('이상치 데이터의 여성의 수 :', len(out_df[out_df['Sex'] == 'female']))
```

## T1-2.Outlier(age) / #이상치 #소수점나이

### 이상치를 찾아라(소수점 나이)
- 데이터셋 : basic1.csv
- 이상치(소수점 나이)를 찾고 올림, 내림, 버림(절사)했을때 3가지 모두 이상치 'age' 평균을 구한 다음 모두 더하여 출력
- https://www.kaggle.com/code/agileteam/py-t1-2-expected-questions/notebook

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/basic1.csv')

df.head(3)
```

```python
out_data = df[(df['age'] - np.floor(df['age'])) != 0]
out_data
```

```python
ceil = np.ceil(out_data['age']).mean() # 올림
floor = np.floor(out_data['age']).mean() # 내림
trunc = np.trunc(out_data['age']).mean() # 버림

print(ceil, floor, trunc)
print(ceil+floor+trunc)
```

## T1-3. Missing data / #결측치 #삭제 #중앙 #평균

### 결측치 처리
- 데이터셋 : basic1.csv
- 결측치가 80%이상 되는 컬럼은(변수는) 삭제, 80% 미만인 결측치가 있는 컬럼은 'city'별 중앙값으로 값을 대체하고 'f1'컬럼의 평균값을 출력
- https://www.kaggle.com/code/agileteam/py-t1-3-map-expected-questions/notebook

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/basic1.csv')

display(df.head())
df.shape
```

```python
df.isnull().sum() / len(df)
```

```python
df.drop('f3', axis=1, inplace=True)
df.shape
```

```python
null_index = df[df['f1'].isnull()].index
df.loc[null_index]
```

```python
df['city'].value_counts()
```

```python
k = df[df['city'] == '경기']['f1'].median()
d = df[df['city'] == '대구']['f1'].median()
s = df[df['city'] == '서울']['f1'].median()
b = df[df['city'] == '부산']['f1'].median()

k, d, s, b
```

```python
df['f1'] = df['f1'].fillna(df['city'].map({'경기':k, '대구':d, '서울':s, '부산':b}))

df.loc[null_index]
```

```python
df['f1'].mean()
```

## T1-4. Skewness and Kurtosis (Log Scale) / #왜도 #첨도 #로그스케일

### 왜도와 첨도 구하기
- 데이터셋 : House Prices_train.csv
- 'SalePrice'컬럼의 왜도와 첨도를 구한 값과, 'SalePrice'컬럼을 스케일링(log1p)로 변환한 이후 왜도와 첨도를 구해 모두 더한 다음 소수점 2째자리까지 출력
- https://www.kaggle.com/code/agileteam/py-t1-4-expected-questions/notebook

> 왜도와 첨도
- 왜도 : 실수값 확률변수의 확률분포 비대칭성 나타내는 지표
- 첨도 : 확률분포의 꼬리가 두꺼운 정도는 나라내는 척도

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/House Prices_train.csv')

display(df)
```

```python
skew_1 = df['SalePrice'].skew() # 왜도
kurt_1 = df['SalePrice'].kurt() # 첨도

skew_1, kurt_1
```

```python
df['SalePrice'] = np.log1p(df['SalePrice']) # 로그 변환

skew_2 = df['SalePrice'].skew()
kurt_2 = df['SalePrice'].kurt()

skew_2, kurt_2
```

```python
print(round(skew_1+skew_2+kurt_1+kurt_2, 2)) # 소수점 둘째자리
```

## T1-6. Groupby Sum / #결측치 #조건

### 결측치 제거 및 그룹 합계에서 조건에 맞는 값 찾아 출력
- 데이터셋 : basic1.csv
- 'f1'컬럼 결측 데이터를 제거하고, 'city'와 'f2'을 기준으로 묶어 합계를 구하고, 'city가 경기이면서 f2가 0'인 조건에 만족하는 f1 값을 구하시오
- https://www.kaggle.com/code/agileteam/py-t1-6-expected-questions/notebook

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./data/basic1.csv')

display(df.head())
df.shape
```

```python
print(df['f1'].isnull().sum())

df[df['f1'].isnull()]
```

```python
df = df[~df['f1'].isnull()]
df[df['f1'].isnull()]
```

```python
df_groupby = df.groupby(['city','f2']).sum().reset_index()
df_groupby[(df_groupby['city'] == '경기') & (df_groupby['f2'] == 0)]['f1']
```

    - 왜도 : df.skew()

    - 로그변환 : np.log()

    - 표준편차 : df.std()
